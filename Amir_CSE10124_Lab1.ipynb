### Task 01-2: Description (0 pts.)
#### Counting and merging repetitive pairs

Clearly there are many repeating pairs of tokens in this list, particularly:
`(117, 103)` and `(117, 110)` which are the character-tokens for `ug` and `un`.

Byte-Pair Encoding works by repeatedly finding these pairs and then combining them into a new token!

In the code cell below, finish the two functions to:
1. Find and count pairs of tokens
2. Combine the target pair into a new token

Then run the two functions on our prompt from above, printing out the length of the prompt before and after running the functions

### Task 01-2: Code (0.5 pts.)

````python
from collections import Counter

def count_pairs(tokens):
    pair_counts = Counter()

    # TODO: Finish the count_pairs function
    # LINES: 2

    for pair in zip(tokens, tokens[1:]):

      # zip stitches 2 lists together creating sort of list of lists
      # tokens[1:] is called slicing: so tokens[1:] starts from index 1
      # tokens[start:stop]

        pair_counts[pair] += 1

    return pair_counts

def merge_pair(tokens, target_pair, combined_id):
    merged_tokens = []

    # TODO: Finish the merge_pair function
    # LINES: 8

    i = 0 # to track where we are in the list
    while i < len(tokens):
      if i < len(tokens) - 1 and (tokens[i], tokens[i+1]) == target_pair:
        merged_tokens.append(combined_id)
        i += 2
      else:
        merged_tokens.append(tokens[i])
        i += 1

    return merged_tokens

'''
overview:
1. count_pairs just counts the number of all consequitive pairs, we will use most_common pair in pair_counts
2. merge_pair goes through the list of tokens and then just merges the 2 tokens that are the most common pair
'''
````

## Task 02: Simple BPE Tokenizer Class
### Task 02-1: Description (0 pts.)
#### Simple BPE Tokenizer Class

In the cell below, fill out the code to complete the class for our simple tokenizer.

### Task 02-1: Code (1 pt.)

````python
"""
Minimal (byte-level) Byte Pair Encoding tokenizer.

Algorithmically follows along the GPT tokenizer:
https://github.com/openai/gpt-2/blob/master/src/encoder.py
"""

class Simple_Tokenizer():
    def __init__(self):
        self.merges = {} # (int, int) -> int

        self.special_tokens = {'<|sos|>': 256, '<|eos|>': 257}

        self.vocab = {idx: bytes([idx]) for idx in range(256)}

        self.vocab[256] = '<|sos|>'
        self.vocab[257] = '<|eos|>'


    def _count_pairs(self, tokens, counts=None):
        pair_counts = Counter()

        # TODO: Finish the _count_pairs function
        # LINES: 2
        for pair in zip(tokens, tokens[1:]):
          pair_counts[pair] += 1


        return pair_counts


    def _merge_pair(self, tokens, target_pair, combined_id):
        merged_tokens = []

        # TODO: Finish the _merge_pair function
        # LINES: 8
        i = 0 # to track where we are in the list
        while i < len(tokens):
          if i < len(tokens) - 1 and (tokens[i], tokens[i+1]) == target_pair:
            merged_tokens.append(combined_id)
            i += 2
          else:
            merged_tokens.append(tokens[i])
            i += 1

        return merged_tokens

    def train(self, text, max_vocab_size=1024):
        vocab_size = len(self.vocab)
        num_merges = max_vocab_size - vocab_size

        text_bytes = text.encode('utf-8')
        tokens = list(text_bytes)

        for i in range(num_merges):
            # TODO: Count the pairs
            pairs = self._count_pairs(tokens)

            # This code will handle us having no more pairs to merge
            try:
                target_pair = pairs.most_common()[0][0]
            except:
                break

            # TODO: Calculate the id for our new pair
            combined_id = vocab_size + i
            # TODO: Merge the pair with the new id
            tokens = self._merge_pair(tokens, target_pair, combined_id)
            # TODO: Save the id for the new pair in self.merges {pair -> id}
            self.merges[target_pair] = combined_id
            # TODO: Save what the id represents in self.vocab {id -> str}
            self.vocab[combined_id] = self.vocab[target_pair[0]] + self.vocab[target_pair[1]]

    def decode(self, tokens):
        text_bytes = b''.join(self.vocab[token] for token in tokens)

        return text_bytes.decode('utf-8')

    def encode(self, text):
        text_bytes = text.encode('utf-8')
        tokens = list(text_bytes)

        while len(tokens) >= 2:
            pairs = self._count_pairs(tokens)

            earliest_pair = min(pairs, key=lambda p: self.merges.get(p, float("inf")))

            if earliest_pair not in self.merges:
                break

            pair_token_id = self.merges[earliest_pair]
            tokens = self._merge_pair(tokens, earliest_pair, pair_token_id)

        return tokens
````

### Task 03-2: Description (0 pts.)
#### Regex Tokenizer Class

In the cell below, fill out the sections marked to finish the `Regex_Tokenizer` class

### Task 03-2: Code (1 pt.)

````python
class Regex_Tokenizer():
    GPT4_SPLIT_PATTERN = r"""'(?i:[sdmt]|ll|ve|re)|[^\r\n\p{L}\p{N}]?+\p{L}+|\p{N}{1,3}| ?[^\s\p{L}\p{N}]++[\r\n]*|\s*[\r\n]|\s+(?!\S)|\s+"""

    def __init__(self):
        self.merges = {} # (int, int) -> int
        self.special_tokens = {'<|sos|>': 256, '<|eos|>': 257} # str -> int, e.g. {'<|endoftext|>': 100257}
        self.vocab = {idx: bytes([idx]) for idx in range(256)} | {idx: special.encode("utf-8") for special, idx in self.special_tokens.items()}

        self.pattern = self.GPT4_SPLIT_PATTERN
        self.compiled_pattern = re.compile(self.pattern)

    def _count_pairs(self, tokens, pair_counts=None):
        for pair in zip(tokens, tokens[1:]):
            pair_counts[pair] += 1

        return pair_counts

    def _merge_pairs(self, tokens, target_pair, combined_id):
        merged_tokens = []

        i = 0
        while i < len(tokens):
            if i < len(tokens) - 1 and (tokens[i], tokens[i+1]) == target_pair:
                merged_tokens.append(combined_id)
                i += 2
            else:
                merged_tokens.append(tokens[i])
                i += 1

        return merged_tokens

    def chunk(self, text):
        return re.findall(self.compiled_pattern, text)

    def train(self, text, max_vocab_size, verbose=False):
        pretrain_vocab_size = len(self.vocab)
        num_merges = max_vocab_size - pretrain_vocab_size

        text_chunks = re.findall(self.compiled_pattern, text)

        ids = [list(ch.encode("utf-8")) for ch in text_chunks]

        for i in range(num_merges):
            pairs = Counter()

            # TODO: for each chunk, count the number of times every consecutive pair appears
            # LINES: 2
            for chunk_ids in ids:
              pairs = self._count_pairs(chunk_ids, pairs)

            pair = pairs.most_common()[0][0]

            idx = pretrain_vocab_size + i

            ids = [self._merge_pairs(chunk_ids, pair, idx) for chunk_ids in ids]

            self.merges[pair] = idx
            self.vocab[idx] = self.vocab[pair[0]] + self.vocab[pair[1]]

    def decode(self, ids):
        # given ids (list of integers), return Python string
        part_bytes = []
        for idx in ids:
            if idx in self.vocab:
                part_bytes.append(self.vocab[idx])
            else:
                part_bytes.append('<|unk|>')

        text_bytes = b"".join(part_bytes)
        text = text_bytes.decode("utf-8", errors="replace")

        return text

    def _encode_chunk(self, text_bytes):
        # return the token ids
        ids = list(text_bytes)

        while len(ids) >= 2:
            # find the pair with the lowest merge index
            stats = self._count_pairs(ids, Counter())
            pair = min(stats, key=lambda p: self.merges.get(p, float("inf")))

            if pair not in self.merges:
                break # nothing else can be merged anymore

            idx = self.merges[pair]
            ids = self._merge_pairs(ids, pair, idx)

        return ids

    def encode(self, text):
        special_pattern = "(" + "|".join(re.escape(k) for k in self.special_tokens) + ")"
        special_chunks = re.split(special_pattern, text)

        tokens = []
        for part in special_chunks[1:-1]:
            if part in self.special_tokens:
                tokens.append(self.special_tokens[part])
            else:
                text_chunks = self.chunk(part)

                # TODO: encode each chunk of text and add the tokens to the list
                # LINES: 4
                # NOTE: You can use the _encode_chunk function above to help with this!
                for chunk in text_chunks:
                  chunk_bytes = chunk.encode('utf-8')
                  tokens.append(self._encode_chunk(chunk_bytes))

        return tokens
````
